```python
from datetime import datetime
from urllib.request import Request, urlopen


def save_url_content(url: str) -> str:
    # URLからデータ取得（403対策としてUser-Agentを付与）
    req = Request(url, headers={"User-Agent": "Mozilla/5.0"})
    with urlopen(req) as response:
        data = response.read()

    # UTF-8でデコードして先頭500文字を取得
    text = data.decode("utf-8", errors="replace")
    head_500 = text[:500]

    # 日時をファイル名にして保存（例: 20241030_155221.txt）
    file_name = datetime.now().strftime("%Y%m%d_%H%M%S") + ".txt"
    with open(file_name, "w", encoding="utf-8") as f:
        f.write(head_500)

    # 作成したファイル名を返す
    return file_name


# 呼び出し側
url = input("URLを入力してください: ")
file_name = save_url_content(url)
print(f"Content saved in: {file_name}")
```
