```python
def is_palindrome(s):
    if s == s[::-1]:
        print(f"「{s}」は回文です！")
    else:
        print(f"「{s}」は回文ではないです。")

is_palindrome("にんてんどううどんてんに")

is_palindrome("Pythonは便利だが遅い")

is_palindrome("no devil lived on")
```

```python
def greet(name):
    print(f"こんにちは、{name}さん！")

greet("太郎")
greet("花子")

```

```python
def introduce(name, age):
    print(f"こんにちは、{name}さん！ 年齢は{age}歳ですね。")

introduce("太郎", 20)
introduce("花子", 25)
```

```python
def introduce(name, age=18):
    print(f"こんにちは、{name}さん！ 年齢は{age}歳ですね。")

introduce("太郎")
introduce("花子", 25)
```

```python
def add(*args):
    result = sum(args)
    print(f"合計:{result}")

add(1, 2, 3)
add(5, 10, 15, 20)
```

```python
def introduce(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} : {value}")

introduce(name="太郎", age=20, hobby="読書")
```

```python
def add(a, b):
    result = a + b
    return result

total = add(3, 5)
print(total)
```
