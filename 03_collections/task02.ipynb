```python
# 1. 英単語を格納するための空のリスト
words: list[str] = []

# 2. アルファベットごとの数を格納する辞書（a～zを0で初期化）
letter_counts = {chr(ord('a') + i): 0 for i in range(26)}

# 3. 英単語の入力受付を繰り返す
while True:
    word = input("英単語を入力してください：").strip()
    if word == "":
        break

    # 英単語をリストに格納（そのまま保持）
    words.append(word)

    # 文字を1つずつ確認してカウント（英字のみ、大小は小文字に統一）
    for ch in word.lower():
        if 'a' <= ch <= 'z':
            letter_counts[ch] += 1

# 4. 英単語をABC順に並べ替える（大文字小文字を無視してソート）
words_sorted = sorted(words, key=str.lower)

# 5. 英単語の一覧表示
print(f"入力した英単語：{words_sorted}")

# 6. 数が0より大きい文字だけ表示
for letter in sorted(letter_counts.keys()):
    count = letter_counts[letter]
    if count > 0:
        print(f"{letter}が{count}個ありました")
```
