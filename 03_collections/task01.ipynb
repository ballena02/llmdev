```python
# 成績リスト（指定のソースコードを使用）
grades = [85, 90, 78, 92, 88]

# 平均点を計算
average = sum(grades) / len(grades)

# 最高点・最低点を取得
max_grade = max(grades)
min_grade = min(grades)

# 平均以上の学生数をカウント
count_above_or_equal_avg = sum(1 for g in grades if g >= average)

# 結果を出力
print(f"平均点: {average:.1f}")
print(f"最高点: {max_grade}")
print(f"最低点: {min_grade}")
print(f"平均以上の学生数: {count_above_or_equal_avg}")
```
