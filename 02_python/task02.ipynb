```python
# 税抜価格を入力
price_ex_tax = int(input("税抜価格を入力してください（円）: "))

# 消費税率
TAX_RATE = 0.10

# 税込価格を計算（小数は四捨五入して整数円にする）
price_in_tax = round(price_ex_tax * (1 + TAX_RATE))

# 送料を決定（送料無料判定は税込価格が2000円以上）
if price_in_tax >= 2000:
    shipping = 0
    print("送料は無料です")
else:
    shipping = 350
    print(f"送料は{shipping}円です")

# 送料込税込価格を計算
total = price_in_tax + shipping

# 結果を表示
print(f"送料込税込価格は{total}円です")
```
